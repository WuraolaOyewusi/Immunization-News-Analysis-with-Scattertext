```
#Subset the dataframe per Label Class
child_df = immunization_df[immunization_df['Label'] == 'Child']
adult_df = immunization_df[immunization_df['Label'] == 'Adult']
```

```
adult_text = adult_df['Headline_teaser'].to_string()
wordcloud = WordCloud(scale=5).generate(adult_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
wordcloud.to_file('adult_immunization_text.jpeg')
```

```
child_text = child_df['Headline_teaser'].to_string()
wordcloud = WordCloud(scale=5).generate(child_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
wordcloud.to_file('child_immunization_text.jpeg')
```

```
nlp = st.WhitespaceNLP.whitespace_nlp_with_sentences
corpus = st.CorpusFromPandas(immunization_df,category_col='Label',text_col='Headline_teaser', nlp=nlp).build()
html = st.produce_scattertext_explorer(corpus,category='Adult',category_name='Adult_Immunization',not_category_name='Child_immunization',width_in_pixels=1210,save_svg_button=True)
```

```
display(HTML(html))
```

```
 #Save as html file
 open("immunization_scattertext.html", 'wb').write(html.encode('utf-8'))
```
